```python
def g(x,y):
    if x==y:
        return "fresher"
    elif x-y==-1:
        return "one year"
    elif x-y==-2:
        return "2 year"
    else:
        return "more than 2 years"
```

```python
df=df.drop(["sd"],axis=1)
```

```python
df["joiningyear"]=df["DOJ"].dt.year
```

```python
df["sd"]=df.apply(lambda x : g(x["GraduationYear"],x["joiningyear"]),axis=1)
```
